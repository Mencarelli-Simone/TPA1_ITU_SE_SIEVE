Spain has specified possible conflicts with certain operators; this script is to find the satellite names corresponding to those operators.

```python
import pandas as pd
import numpy as np
import os
from IPython.display import display
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
pd.set_option('display.max_rows', None)
```

```python
database_file = './countriestables/E.csv'

database = pd.read_csv(database_file)

display(database.head())
print(database.columns)
```

```python
# find all names containing the word HISPASAT and EMAD
operatornames = ['HISPASAT','EMAD']
names = database[' com_el.sat_name'].unique().tolist()
operatorsats = []
for operator in operatornames:
    for name in names:
        if operator in name:
            operatorsats.append(name)
            
for op in operatorsats: print(op)
```
